```python
def read_results(results_directory, path = False):
    #Read in all 'reactions.tsv' from the subdirectories in the compass_results folder.
    if not path: 
        compass_results_dir = os.path.join('/home/dzhang/Compass_Results/HumanGEM_PANC/full_compass_results', results_directory) 
    else:
        compass_results_dir = results_directory
    
    # Get the list of directories in compass_results
    directories = [folder for folder in os.listdir(compass_results_dir) if os.path.isdir(os.path.join(compass_results_dir, folder))]
    
    # Create a dictionary to store the penalties DataFrames
    scores_dict = {}

    # Iterate over the directories
    for directory in directories:
        if not directory.startswith('_tmp_'):
            scores_file = os.path.join(compass_results_dir, directory, 'reactions.tsv')
        
            # Read the penalties.csv file if it exists
            if os.path.isfile(scores_file):
                scores_df = pd.read_csv(scores_file, sep="\t", index_col = 0)
            
                # Store the penalties DataFrame in the dictionary
                scores_dict[directory] = scores_df
                print(f"Run complete: Reactions.tsv found in directory {directory}.")
            # else:
            #     print(f"No reactions.tsv file found in directory {directory}.")
    return scores_dict
```

```python
def read_metadata(files_directory, path = False):
    #Read in all 'metadata.tsv' from the compass_files folder.
    if not path:
        compass_files_dir = '/home/dzhang/Compass_Results/PANC_scRNAseq/compass_files/'+files_directory
    else: 
        compass_files_dir = files_directory 
    metadata_dict = {}

    # Get the list of directories in compass_files
    directories = [folder for folder in os.listdir(compass_files_dir) if os.path.isdir(os.path.join(compass_files_dir, folder))]
   
    # Iterate over the files in compass_files directory
    for directory in directories: 
        directory_file = os.path.join(compass_files_dir, directory) 
        for filename in os.listdir(directory_file):
            if filename.endswith('_metadata.tsv'):
                # Get the file prefix by removing the extension and splitting by '_'
                file_prefix = os.path.splitext(filename)[0].split('_', 1)[0]
                file_suffix = os.path.splitext(filename)[0].split('_', 1)[1].rsplit('_', 1)[0]
                key_rename = file_prefix+"_"+file_suffix+"_"+directory
                file_path = os.path.join(directory_file, filename)
            
                # Read the metadata file and store in the dictionary
                metadata_dict[key_rename] = pd.read_csv(file_path, sep='\t')

    return metadata_dict
```

```python
def welcht_test(consistencies_matrix, group_A_cells, group_B_cells):
    """
    Performs Welch's t-test for each reaction between group_A and group_B
    """
    #per reaction/meta-reaction, conduct wilcoxon test between group_A and group_B
    group_A = consistencies_matrix.loc[:,group_A_cells]
    group_B = consistencies_matrix.loc[:,group_B_cells]
    results = pd.DataFrame(index = consistencies_matrix.index, columns = ['welcht_stat', 'welcht_pval', 'cohens_d'], dtype='float64')
    for rxn in consistencies_matrix.index:
        A, B = group_A.loc[rxn].to_numpy().ravel(), group_B.loc[rxn].to_numpy().ravel()
        #sometimes there's a solitary value, and we don't want to test then
        if len(np.unique(A)) == 1 and len(np.unique(B)) == 1:
            if np.unique(A) == np.unique(B):
                #we've got no data. set p-value to 1 and skip!
                #(p-value needs to be 1 so multipletests doesn't cry)
                results.loc[rxn, ['welcht_pval']] = 1
                continue
        stat, pval = ttest_ind(A, B, equal_var=False, alternative='two-sided')
        c_d = cohens_d(A, B)
        results.loc[rxn, ['welcht_stat', 'welcht_pval', 'cohens_d']] = stat, pval, c_d
    results['adjusted_pval'] = np.array(multipletests(results['welcht_pval'], method='fdr_bh')[1], dtype='float64')
    return results
```

```python
def penalties2rxnscores(penalties_dict, reaction_metadata, comparison_keys, test_type):
    #penalties_dict should be created by read_results
    #comparison_keys should be a list of two keys that are in penalties_dict
    #test_type: 'ttest' or 'wilcox'
    rxnconsistencies_dict = {}
    for directory, penalties_df in penalties_dict.items():
        rxnconsistencies_dict[directory] = get_reaction_consistencies(penalties_df)
    reaction_consistencies = rxnconsistencies_dict[comparison_keys[0]].merge(rxnconsistencies_dict[comparison_keys[1]],left_index=True, right_index = True)
    if test_type == 'wilcox':
        w_results = wilcoxon_test(reaction_consistencies, rxnconsistencies_dict[comparison_keys[0]].T.index.to_list(), rxnconsistencies_dict[comparison_keys[1]].T.index.to_list())
    else:
        w_results = welcht_test(reaction_consistencies, rxnconsistencies_dict[comparison_keys[0]].T.index.to_list(), rxnconsistencies_dict[comparison_keys[1]].T.index.to_list())    
                                                                                                                                                 
    w_results['metadata_r_id'] = ""
    for r in w_results.index:
        if r in reaction_metadata.index:
            w_results.loc[r, 'metadata_r_id'] = r
        elif r[:-4] in reaction_metadata.index:
            w_results.loc[r, 'metadata_r_id'] = r[:-4]
        else:
            print(f"{r} does not occur in metadata")
            w_results.loc[r, 'metadata_r_id'] = r[:-4]

    W = w_results.merge(reaction_metadata, how='left', left_on='metadata_r_id', right_index=True, validate='m:1')
        
    W_rest = W[~W['CONFIDENCE SCORE'].isin([0,4])]
    W_rest = W_rest[~W_rest['EC-NUMBER'].isna()]

    # 13036 rxns have 0
    # 37 have 2
    # 12 have 4
    W = W[W['CONFIDENCE SCORE'].isin([0,4])]
    W = W[~W['EC-NUMBER'].isna()]
    # print('W size=', W.shape, ', Wilcox_result size = ', w_results.shape)
              
    return W, W_rest, reaction_consistencies
```

```python
def trim_rxn_labels(full_label_dict):
    # trim rxn enzyme names shorter for display in the plot
    # used in plots that add rxn labels, e.g. compare_plot_differential_scores_labels
    temp = []
    unique_rxn_labels = dict()
    for key, val in full_label_dict.items():
        if val not in temp:
            temp.append(val)
            if len(val) > 40:
                result = re.split('ase', val)
                short_val = result[0]+'ase'
                unique_rxn_labels[key] = short_val
            else:
                unique_rxn_labels[key] = val
                
    return unique_rxn_labels           
```

```python
def trim_rxn_genes(full_label_dict):
    # trim rxn gene list shorter for display in the plot
    # used in plots that add rxn labels, e.g. compare_plot_differential_scores_labels
    temp = []
    unique_rxn_labels = dict()
    for key, val in full_label_dict.items():
        if val not in temp:
            temp.append(val)
            if type(val)==str:
                # exclude rxns associated with no genes (val is nan)
                val_split = val.split(";")
                if len(val_split) > 3:
                    short_val = ';'.join(val_split[0:2])
                    unique_rxn_labels[key] = short_val
                else:
                    unique_rxn_labels[key] = val
                
    return unique_rxn_labels  
```

```python
def summary_plot(rxnscores, comparison_keys, title = 'Reaction scores in subsystems', p_cutoff = 0.05, d_yaxref = None, xlim = [-10,10]):
    # rxnscores: the output of penalties2rxnscores
    # comparison_keys should be a list of two keys that are in penalties_dict, same as in penalties2rxnscores
    # p_cutoff: value of cutoff p value for the transprency for dots color
    # d_yaxref: None or reference axis name, ususally from a previous plot
    
    W = rxnscores[0]
    data = W[~W['SUBSYSTEM'].isin(["Miscellaneous", "Unassigned"])]
    data = data[~data['SUBSYSTEM'].map(lambda x: "Transport" in x or "Exchange" in x or x == "Other")]
    items, counts = np.unique(data['SUBSYSTEM'], return_counts=True)
    items = [items[i] for i in range(len(items)) if counts[i] > 3] # filter(n() > 5)  used in the paper
    data = data[data['SUBSYSTEM'].isin(items)]
    
    right_color = '#FF0000' # red
    left_color =  '#0000FF' # blue
    
    fig, axs = plt.subplots(figsize = (8, 13))
    
    # Either take in a reference for the y-axis or sort as default
    if d_yaxref is None:
        d = data[data['adjusted_pval'] < 0.15].groupby('SUBSYSTEM')['cohens_d'].median()
        axs.scatter(d[d.argsort], d[d.argsort].index, alpha=0)
    else:
        d = data[data['adjusted_pval'] < 0.15].groupby('SUBSYSTEM')['cohens_d'].median()
        axs.scatter(d_yaxref[d_yaxref.argsort], d_yaxref[d_yaxref.argsort].index, alpha=0)
    
    color = data['cohens_d'].map(lambda x: right_color if x >= 0 else left_color)
    alpha = data['adjusted_pval'].map(lambda x: 0.75 if x < p_cutoff else 0.1)
    axs.scatter(data['cohens_d'], data['SUBSYSTEM'], s = 10, c=color, alpha=alpha)
    axs.axvline(0, dashes=(3,3), c='black')
    axs.grid(axis = 'y', color = 'gray', linestyle = '--', linewidth = 0.5)
    axs.set_xlabel("Cohen's d", fontsize=12)
    axs.tick_params(axis='y', which='major', labelsize=8)
    axs.set_title(title, fontdict={'fontsize':16})
    axs.annotate(comparison_keys[0], xy=(0.7, -0.08), xycoords='axes fraction', color=right_color, fontsize=12)
    axs.annotate(comparison_keys[1], xy=(-0.1, -0.08), xycoords='axes fraction', color=left_color, fontsize=12)
    
    axs.set_xlim(xlim[0], xlim[1])
    
    plt.show()
    
    return d # d saves current axis info, used for the reference axis
```

```python
def multiple_summary_plots(penalties_dict, reaction_metadata, comparison_keys_list, test_type = 'wilcox', title = 'Reaction scores in subsystems', p_cutoff = 0.05, d_yaxref = None, xlim = [-10,10]):
    #penalties_dict should be created by read_results
    #comparison_keys_list: list of pairs of comparison keys in the format: [[a,b],[c,d]]
    for i, comparison_keys in enumerate(comparison_keys_list):
        if i == 0:
            rxnscores = penalties2rxnscores(penalties_dict, reaction_metadata, comparison_keys, test_type)
            ref_plot = summary_plot(rxnscores, comparison_keys, title, p_cutoff, d_yaxref, xlim)
        else:
            rxnscores = penalties2rxnscores(penalties_dict, reaction_metadata, comparison_keys, test_type)
            d_plot = summary_plot(rxnscores, comparison_keys, title, p_cutoff, ref_plot, xlim)
```

```python
def volcano_plot(rxnscores, comparison_keys, subsystem, gene_or_enz, title = 'Subsystem scores', p_cutoff = 0.05, xlim = [-10,10]):
    """
    rxnscores: the output of penalties2rxnscores
    comparison_keys should be a list of two keys that are in penalties_dict, same as in penalties2rxnscores
    subsystem: subsystem in rxnscores, ex 'Citric acid cycle'
    p_cutoff: p-value for wilcoxon test
    gene_or_enz: labels that will appear by the dots; 'gene' or 'enzyme'
    """
    
    W = rxnscores[0]
    data = W[W['SUBSYSTEM'] == subsystem]
    subsystem_data = data[data['adjusted_pval'] < p_cutoff].sort_values(by = ['adjusted_pval', 'cohens_d'], ascending = [True, False])
    n_labels = len(subsystem_data)
    
    if gene_or_enz.lower() == 'enzyme':
        top_reactions = subsystem_data.iloc[0:n_labels, 5].to_dict()
        labeled_rxns = trim_rxn_labels(top_reactions)
    elif gene_or_enz.lower() == 'gene':
        top_reactions = subsystem_data["GENE NAME ASSOCIATION"].to_dict()
        labeled_rxns = trim_rxn_genes(top_reactions)
    elif gene_or_enz.lower() == 'name':
        labeled_rxns = subsystem_data['NAME'].to_dict() 
        
    print(labeled_rxns)
              
    plt.figure(figsize=(5,6))
    axs = plt.gca()
    
    right_color = '#FF0000' # red
    left_color =  '#0000FF' # blue 
    
    set_color = data['cohens_d'].map(lambda x: right_color if x >= 0 else left_color)
    set_alpha = data['adjusted_pval'].map(lambda x: 0.75 if x < 0.05 else 0.1)
    
    axs.scatter(data['cohens_d'], -np.log10(data['adjusted_pval']), c=set_color, s=20,  alpha=set_alpha)
    axs.set_xlabel("Cohen's d", fontsize=8)
    axs.set_ylabel("-log10 (Adjusted p)", fontsize=8)
    
    #Everything after this should be tweaked depending on your application
    axs.set_xlim(xlim[0], xlim[1])
    axs.axvline(0, dashes=(), c='gray')
    axs.axhline(1.3, dashes=(3,6), c='black')
    axs.axhline(2, dashes=(3,6), c='gray')
    
    axs.set_title(subsystem, fontdict={'fontsize':10})
    axs.annotate(comparison_keys[0], xy=(0.6, -0.18), xycoords='axes fraction', color=right_color,fontsize=8)
    axs.annotate(comparison_keys[1], xy=(-0.1, -0.18), xycoords='axes fraction', color=left_color,fontsize=8)
    
    texts = [] 
    for r, label in labeled_rxns.items():
        x = data.loc[r, 'cohens_d']
        y = -np.log10(data.loc[r, 'adjusted_pval'])
        if x > 0:
            offset = (20, 0)
            text = axs.annotate(label, (x,y), xytext = offset, color = 'black', fontsize = 2,
                                textcoords='offset pixels', arrowprops={'width': 0.001, 'headwidth':0.1, 'headlength':0.2, 'shrink': 0.05})
        elif x < 0:
            offset = (-40, -20)
            text = axs.annotate(label, (x,y), xytext = offset, color = 'black', fontsize = 2,
                                textcoords='offset pixels', arrowprops={'width': 0.001, 'headwidth':0.1, 'headlength':0.2, 'shrink': 0.05})
        texts.append(text)  # Add text annotation to the list
    if texts:  # Check if there are text annotations
        adjust_text(texts, ax=axs)
```

```python
def save_significant_rxns(rxnscores, comparison_keys, results_directory, p_cutoff = 0.05):
    # save the rxns that are significantly diffferent in two comparing sets (adj_pval < p_cutoff)
    # rxnscores: the output of penalties2rxnscores
    # comparison_keys should be a list of two keys that are in penalties_dict, same as in penalties2rxnscores
    # results_directory is the folder within compass_results from which this comparison is derived, ex single_sample_L1
    
        W = rxnscores[0]
        W_pcut = W[W["adjusted_pval"] < p_cutoff].copy()
        W_pcut['abs_cohens_d'] = np.abs(W_pcut['cohens_d'])
        W_pcut_sort = pd.DataFrame(columns = W_pcut.columns.to_list())

        #sort each subsystem by absolute cohen's d and reappend
        for subsystem in np.sort(W.subsystem.unique()):
            W_subsystem_sorted = W_pcut[W_pcut['SUBSYSTEM'] == subsystem].sort_values(
                                        by = ['adjusted_pval', 'abs_cohens_d'], ascending = [True, False])
            W_pcut_sort = pd.concat([W_pcut_sort, W_subsystem_sorted], sort=False)

        # Create output directory if it doesn't exist
        compass_results_dir = 'compass_results/'+results_directory
        output_path = os.path.join(compass_results_dir, "significant_genes/")
        os.makedirs(output_path, exist_ok=True)

        # Generate output name based on comparison keys and subsystem
        output_name = f"{comparison_keys[0]}_{comparison_keys[1]}_{subsystem}.csv"
        output_file = os.path.join(output_path, output_name)

        W_pcut_sort.to_csv(output_file, index=True)
```

```python
def multiple_save_significant_rxns(penalties_dict, reaction_metadata, comparison_keys_list, results_directory, p_cutoff = 0.05):
    # save the rxns that are significantly diffferent in two comparing sets (adj_pval < p_cutoff)
    # rxnscores: the output of penalties2rxnscores
    # comparison_keys should be a list of two keys that are in penalties_dict, same as in penalties2rxnscores
    # results_directory is the folder within compass_results from which this comparison is derived, ex single_sample_L1
    for i, comparison_keys in enumerate(comparison_keys_list):
        
        rxnscores = penalties2rxnscores(penalties_dict, reaction_metadata, comparison_keys, 'wilcox', reaction_metadata)
        
        W = rxnscores[0]
        W_pcut = W[W["adjusted_pval"] < p_cutoff].copy()
        W_pcut['abs_cohens_d'] = np.abs(W_pcut['cohens_d'])
        W_pcut_sort = pd.DataFrame(columns = W_pcut.columns.to_list())

        #sort each subsystem by absolute cohen's d and reappend
        for subsystem in np.sort(W.SUBSYSTEM.unique()):
            W_subsystem_sorted = W_pcut[W_pcut['SUBSYSTEM'] == subsystem].sort_values(
                                        by = ['adjusted_pval', 'abs_cohens_d'], ascending = [True, False])
            W_pcut_sort = pd.concat([W_pcut_sort, W_subsystem_sorted], sort=False)

        # Create output directory if it doesn't exist
        compass_results_dir = '/home/dzhang/Compass_Results/PANC_scRNAseq/compass_results/'+results_directory
        output_path = os.path.join(compass_results_dir, "significant_genes/")
        os.makedirs(output_path, exist_ok=True)

        # Generate output name based on comparison keys and subsystem
        output_name = f"{comparison_keys[0]}_{comparison_keys[1]}_{subsystem}.csv"
        output_file = os.path.join(output_path, output_name)

        W_pcut_sort.to_csv(output_file, index=True)
```

```python
# Identify which subsystems in each model have the most reactions in most common with each other 

associations = {} 
for subsystem in set(HG_metadata['SUBSYSTEM']):
    max = 0 
    for rsubsystem in set(RECON_metadata['subsystem']): 
        # Find max corresponding subsystem in RECON
        count = 0
        for rxnid in RECON_metadata[RECON_metadata['subsystem'] == rsubsystem].index:
            if rxnid in list(HG_metadata[HG_metadata['SUBSYSTEM'] == subsystem]['rxnBiGGID']):
                count += 1
        if count > max: 
            max = count 
            associations[subsystem] = rsubsystem 
associations
```
