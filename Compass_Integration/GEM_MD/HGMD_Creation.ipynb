## HumanGEM Metadata Creation 

Generate a HumanGEM metadata file in accordance with the provided RECON 2 MD file in Compass

```python
from os.path import join
import os
import copy
import scipy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
import random
import re
```

```python
compass_model_PATH = '/home/dzhang/miniconda3/envs/compass_test/lib/python3.7/site-packages/compass/'
reaction_metadata_PATH = join(compass_model_PATH, 'Resources/Recon2_export/rxn_md.csv')
reaction_metadata = pd.read_csv(reaction_metadata_PATH, index_col = 0) 
```

```python
reaction_metadata = reaction_metadata.rename(columns = {"rxn_EC_number":"EC_number", "rxn_confidence":"confidence"})
```

```python
# Load HumanGEM MD file
HumanGEM_metadata_PATH = join('/home/dzhang/Compass_Integration/GEM_MD', 'HG_v115.xlsx')
HumanGEM_metadata = pd.read_excel(HumanGEM_metadata_PATH)
HumanGEM_metadata.set_index('ID', inplace=True)
HumanGEM_metadata = HumanGEM_metadata.drop(['#', 'LOWER BOUND', 'UPPER BOUND', 'OBJECTIVE', 'COMPARTMENT', 'REPLACEMENT ID', 'NOTE', 'REFERENCE'], axis = 1)
```

```python
HumanGEM_metadata
```

```python
# Rearrange column names  
HumanGEM_metadata['GENE NAME ASSOCIATION'] = ""
colnames = HumanGEM_metadata.columns.tolist()
colnames = colnames[:4] + colnames[-1:] + colnames[5:6] + colnames[4:5] + colnames[6:7]
HumanGEM_metadata = HumanGEM_metadata[colnames]  
```

```python
HumanGEM_metadata
```

```python
# Load in HumanGEM model dict to convert IDs 
import json
RESOURCE_DIR = "/home/dzhang/Compass_Integration/Compass/compass/Resources"
hgnc_file = os.path.join(RESOURCE_DIR, "Genes", "new_HGNC.json")

f = open(hgnc_file)

data_dict = json.load(f)
genes = data_dict
```

```python
key_field = "ensembl_gene_id"
gene_dict = {}
for gene in genes: 
    # Not all genes have Ensembl IDs
    try:
        gene_dict[genes[gene][key_field]] = genes[gene]
    except:
        pass 
```

```python
# Parse gene association structure 

for rxn in HumanGEM_metadata.index:
    all_genes = HumanGEM_metadata.loc[rxn]['GENE ASSOCIATION']
    if isinstance(all_genes, str):
        all_genes = all_genes.split("or")
        all_genes = [name.replace(" ", "") for name in all_genes]
        all_genes = [name.replace(")", "") for name in all_genes]
        all_genes = [name.replace("(", "") for name in all_genes]
        
        all_genes = [name.split("and") for name in all_genes]
        
        for group in all_genes:
            # Check for AND association
            if type(group) == list: 
                # Every ENSG maps to one not necessarily unique gene symbol 
                for index, gene in enumerate(group): 
                    group[index] = gene_dict[gene]['symbol']
    HumanGEM_metadata.at[rxn, 'GENE NAME ASSOCIATION'] = all_genes  
```

```python
# Split MIRIAM info into separate IDs and generate column for each 

HumanGEM_metadata['rxnKEGGID'] = ""
HumanGEM_metadata['rxnBiGGID'] = ""
HumanGEM_metadata['rxnRecon3DID'] = ""
HumanGEM_metadata['rxnMetaNetXID'] = ""
HumanGEM_metadata['rxnRheaID'] = ""
HumanGEM_metadata['rxnRheaMasterID'] = ""

# Change col order - bring confidence score to end 
```

```python
# Load rxn name MD file 
reaction_metadata_PATH = os.path.join('/home/dzhang/Compass_Integration/HumanGEM_MD', 'reactions_1.15.tsv')
reaction_metadata = pd.read_csv(reaction_metadata_PATH, sep='\t')
```

```python
reaction_metadata
```

```python
# Assign corresponding rxn IDs
absent_rxn_ids = []
id_types = ["rxnKEGGID", "rxnBiGGID", "rxnRecon3DID", "rxnMetaNetXID", "rxnRheaID",	"rxnRheaMasterID"]
for rxn in HumanGEM_metadata.index:
    # Keep track of which reaction ids are present 
    if reaction_metadata.loc[reaction_metadata['rxns'] == rxn].empty:
        absent_rxn_ids.append(rxn)
    else:
        for id_type in id_types:
            HumanGEM_metadata.at[rxn, id_type] = reaction_metadata.loc[reaction_metadata['rxns'] == rxn][id_type].values[0]
```

```python
absent_rxn_ids
```

```python
HumanGEM_metadata
```

```python
# Parse MIRIAM string for remaining absent rxns to fill in rxn IDs

for rxn in absent_rxn_ids: 
    miriam = HumanGEM_metadata.loc[rxn]['MIRIAM']

    HumanGEM_metadata.at[rxn, "rxnRecon3DID"] = np.nan
    
    # Lack of clear pattern - hard code 
    id_types = ["kegg.reaction", "bigg.reaction", "metanetx.reaction", "rhea"]
    ids = {}
    for id_type in id_types: 
        id = re.findall(rf"{id_type}/\w*;", miriam) 
        if (len(id) == 0):
            ids[id_type] = np.nan 
            
        # Parse id
        # Handle Rhea case separately 
        elif (id_type == "rhea"): 
            rhea = id[0][len(id_type) + 1: -1]
            if (len(id) > 1): 
                # Second rhea ID is master
                master_rhea = id[1][len(id_type) + 1: -1]   
                HumanGEM_metadata.at[rxn, "rxnRheaID"] = f"RHEA:{rhea}"
                HumanGEM_metadata.at[rxn, "rxnRheaMasterID"] = f"RHEA:{master_rhea}"
            else: 
                HumanGEM_metadata.at[rxn, "rxnRheaID"] = np.nan 
                HumanGEM_metadata.at[rxn, "rxnRheaMasterID"] = f"RHEA:{rhea}"
        else: 
            # Regardless of number of different IDs for same ID type, use only first one 
            id = id[0][len(id_type) + 1: -1] 
            ids[id_type] = id
            
        # Append to metadata dataframe 
        for key, value in ids.items(): 
            if key == "kegg.reaction": 
                HumanGEM_metadata.at[rxn, "rxnKEGGID"] = value 
            elif key == "bigg.reaction": 
                HumanGEM_metadata.at[rxn, "rxnBiGGID"] = value 
            elif key == "metanetx.reaction":
                HumanGEM_metadata.at[rxn, "rxnMetaNetXID"] = value 
```
